```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("train_jqd04QH.csv")
test = pd.read_csv("test_GYi4Gz5.csv")
train.head
test.head
```

```python
test['target'] = 0
df = pd.concat([train,test],ignore_index = True)
```

```python
df['city'] = df['city'].map(lambda x: x.lstrip('city_')).astype(int)
```

```python
df['gender'].fillna('Male',inplace = True)
```

```python
df['enrolled_university'].fillna('no_enrollment',inplace = True)
```

```python
df['education_level'].fillna('Graduate', inplace= True)
df['major_discipline'].fillna('STEM',inplace = True)
```

```python
df['experience'].replace('>20','21',inplace= True)
```

```python
df['experience'].replace('<1','0',inplace = True)
```

```python
df['experience'].fillna('21',inplace = True)
df['experience'].astype(int)
mapping = {"<10":0,"10/49":1,"50-99":2,"100-500":3,"500-999":4,"1000-4999":5,"5000-9999":6,"10000+":7}
df['company_size']=df['company_size'].map(mapping)
```

```python
for i in range(len(df)):
    if(df.loc[i,'company_type'] == "Early Stage Startup" and pd.isnull(df.loc[i,'company_size'])):
        df.loc[i,'company_size'] = 0
    elif(df.loc[i,'company_type'] == "Funded Startup" and pd.isnull(df.loc[i,'company_size'])):
        df.loc[i,'company_size'] = 2
    elif(df.loc[i,'company_type'] == "NGO" and pd.isnull(df.loc[i,'company_size'])):
        df.loc[i,'company_size'] = 3
    elif(df.loc[i,'company_type'] == "Other" and pd.isnull(df.loc[i,'company_size'])):
        df.loc[i,'company_size'] = 2
    elif(df.loc[i,'company_type'] == "Public Sector" and pd.isnull(df.loc[i,'company_size'])):
        df.loc[i,'company_size'] = 5
    elif(df.loc[i,'company_type'] == "Pvt Ltd" and pd.isnull(df.loc[i,'company_size'])):
        df.loc[i,'company_size'] = 2
```

```python
df['company_size'].fillna(2,inplace=True)
```

```python
df['company_type'].fillna('Pvt Ltd',inplace = True)
```

```python
mapping2 = {"never":0,"1":1,"2":2,"3":3,"4":4,">4":5}
df['last_new_job'] = df['last_new_job'].map(mapping2)
```

```python
df['last_new_job'].fillna(1,inplace = True)
```

```python
df['experience'] = df['experience'].astype(int)
```

```python
df
```

```python
df.drop('enrollee_id',axis = 1,inplace = True)
```

```python
df1 = df[:18359]
df2 = df[18359:]
```

```python
X_train =df1.drop('target', axis =1).values
Y_train = df1['target'].values
X_test = df2.drop('target', axis =1)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X_train,Y_train,test_size = 0.30,random_state = 10)
```
